## <font color = 'blue'>Validation : prediction of market segments for all customers excluded from data-sampling</font>

```python
import pandas as pd

#---------------------------------------------------------------------
# Get list of all customers out of sample used for M.L train/test
#---------------------------------------------------------------------
listCustomer_out_sample = _oP5_SegmentClassifier.get_listCustomer_out_sample(customerCount=None)
nb_customers = len(listCustomer_out_sample)
nb_customers

dict_customer_cluster = dict()
i_=0
for customerID in listCustomer_out_sample:
    df_customerID = _oP5_SegmentClassifier.get_customer_history_df_invoice_line(customerID)
    cluster = _oP5_SegmentClassifier.get_customer_marketSegment(df_customerID)
    
    if cluster not in dict_customer_cluster.keys():
        list_items = [customerID]
    else:
        list_items = dict_customer_cluster[cluster]
        list_items.append(customerID)    
    
    dict_customer_cluster[cluster] = list_items
    i_ += 1
    print("Processed = "+str(i_)+"/"+str(nb_customers))
    if i_== 3:
        pass
```

```python
for cluster in dict_customer_cluster.keys():
    print("Effective in cluster {} : {}".format(cluster, len(dict_customer_cluster[cluster])))
print("")
for cluster in dict_customer_cluster.keys():
    eff_cluster = len(dict_customer_cluster[cluster])
    print("Cluster "+str(cluster)+" : % = {0:1.2F}".format(100*eff_cluster/nb_customers))
```

### <font color = 'blue'>Validation : prediction of market segments for a RFM score 111 customer</font>

Recency score 1 :   < 15 days from max date / Most recent day from now : 15317

Frequency score 1 : > 5 invoices

Monetary score 1 :  > 1639


```python
import p5_util
#-------------------------------------------------------------------------
# Get F-score from dataframe with data excluded from train/test
#-------------------------------------------------------------------------
self = _oP5_SegmentClassifier
df_freq = p5_util.get_invoiceFreq_from_invoiceLine(self._df_invoice_line_out_sample)
df_freq.sample()

#-------------------------------------------------------------------------
# Get all customers with frequency matching with F=1
#-------------------------------------------------------------------------
ser_freq = df_freq.query('Frequency > 5')
list_df_invoice_line_customer = list()

#-------------------------------------------------------------------------
# Get list of customers with RFM scores = 111
#-------------------------------------------------------------------------
for customerID in ser_freq.index:
    #df_invoice_line_customer = self.get_customer_history_df_invoice_line(customerID)
    df_invoice_line_customer = self._df_invoice_line_out_sample[self._df_invoice_line_out_sample.CustomerID == customerID]
    df, df_RFM, df_RFM_quantiles, day_now = p5_util.p5_df_rfm_build(df_invoice_line_customer\
                                                                    ,df_RFM_threshold=self.df_RFM_quantiles
                                                                    ,day_now = self._day_now)
    RFM_score = df_RFM.loc[0,'RFM']
    if '111' == RFM_score:
        print("customerID= "+str(customerID)+" found with score= "+str(RFM_score))
        list_df_invoice_line_customer.append(df_invoice_line_customer)
        
```

```python
print("Number of customers with RFM 111= "+str(len(list_df_invoice_line_customer)))
```
